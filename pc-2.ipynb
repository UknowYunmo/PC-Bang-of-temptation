```python
import json
import requests
import sys

url = "https://dapi.kakao.com/v2/local/search/keyword.json?"
apikey = "ccf9c4dda0eed674c075b234b8d3d390"
loc=['중곡동','능동','구의동','광장동','자양동','화양동','군자동'] # 광진구
for k in loc:
    pc_list=[]
    place_list=[]
    x_list=[]
    y_list=[]
    phone_list=[]
    print('-------------------------------'+str(k)+'----------------------------------')
    print()
    print()
    query = k+"pc방"
    for i in range(1,11):
        r = requests.get( url, params = {'page':str(i), 'query':query}, headers={'Authorization' : 'KakaoAK ' + apikey } ).text
        r = json.loads(str(r))
        match_first = r['documents']
        for j in match_first :
            if k in j['address_name'] :
                pc_list.append(j['place_name'])
                place_list.append(j['address_name'])
                x_list.append(j['x'])
                y_list.append(j['y'])
                phone_list.append(j['phone'])
    print(pc_list)
    print(place_list)
    print(x_list)
    print(y_list)
    print(phone_list)
    print(len(pc_list))
```
