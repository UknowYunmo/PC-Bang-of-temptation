```python
from folium.features import CustomIcon
m = folium.Map(location=[37.5468565308905, 127.103025712709],zoom_start=15)

icon_image1='C:/Users/user/va.jpg'
icon1 = CustomIcon(
    icon_image1,
    icon_size=(80, 60),
    icon_anchor=(10, 30),
    shadow_image=None,
    shadow_size=None,
    shadow_anchor=None,
    popup_anchor=(30, -30)
)

marker1 = folium.Marker(
    location=[37.5442656969341, 127.098283156877],
    icon=icon1,
    popup=folium.Popup("학교", max_width='40')
)
m.add_child(marker1)
```
